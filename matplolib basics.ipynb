```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('company_sales_data.csv')
obj=pd.DataFrame(df)
obj
```

# 1) Use the company sales data and plot the Total profit of all months. 

```python
x=obj['month_number']
y=obj['total_profit']
a=plt.plot(x,y)
plt.xlabel('months')
plt.ylabel('profits')
plt.title('profits of all months')
```

# 2) Read toothpaste sales data of each month and show it using a scatter plot.
           Also, add a grid in the plot. gridline style should “–“.


```python
month=obj['month_number']
paste_details=obj['toothpaste']
plt.scatter(month,paste_details)
plt.xlabel('months')
plt.ylabel('tooth_paste')
plt.title('details_of_tooth_paste')
plt.grid(linestyle='-')
plt.show()
```

# 3) Plot the Total profit column with Boxplot(horizontal, Vertical, Notch), Histogram, barplot, Dotchart.

```python
profit=obj["total_profit"]
plt.boxplot(profit,vert=0,notch=True)
plt.show()
```

```python
profit=obj["total_profit"]
plt.boxplot(profit,vert=1,notch=True)
plt.show()
```

```python
profit=obj["total_profit"]
months=obj['month_number']
plt.bar(months,profit)
plt.xlabel("months",c='r',size=15)
plt.ylabel("profits",c="r",size=15)
plt.title('company profit details',c="b",size=20)
plt.legend('profits')
plt.show()
```

```python
profit=obj["total_profit"]
months=obj['month_number']
plt.hist(profit)

plt.xlabel("profits",c="r",size=15)
plt.title('company profit details',c="b",size=20)
plt.legend('profits')
plt.show()
```

# 4) Do the scatterplot on Total profit & Month_number column.

```python
profit=obj['total_profit']
month=obj[['month_number']]
plt.scatter(profit,month,c="r")
plt.grid(c="c")
plt.xlabel("profits",size=15,c="g")
plt.ylabel("months",size=15,c="g")
plt.title("month on month profits",size=20,color="blue")
plt.show()
```

# 5)  Plot the Total profit with format” red square -” (“rs-”) and only “red circles “ (“ro”).

```python
profit=obj["total_profit"]
plt.plot(profit,c="r",marker=".",ls="-.")

plt.show()
```
